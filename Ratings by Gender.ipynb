```python
# Import pandas and sqlalchemy
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Resources/IMDb ratings.csv"
```

```python
ratings_data_df = pd.read_csv(csv_file)
```

```python
ratings_data_df.head()
```

```python
# Removed irrelevant columns and reorganized the votes by gender regardless of age.
# The "mean_vote" and "median_vote" columns were also removed due to redundancy. 
# The "total_votes" and "weighted_average_vote" columns were switched around for the sake of clarity.  
Updated_ratings_data_df = ratings_data_df[['imdb_title_id', 'total_votes', 'weighted_average_vote', 'males_allages_avg_vote', 'males_allages_votes', 'females_allages_avg_vote', 'females_allages_votes']].copy()
```

```python
Updated_ratings_data_df.set_index ("imdb_title_id", inplace = True)
```

```python
Updated_ratings_data_df.head()
```

```python
# Connect to database and create engine:
rds_connection_string = "postgres:bootcamp@localhost:5433/Netflix_IMDb_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Check tables
engine.table_names()
```

```python
# Verify that the databased is connected to the table. 
Updated_ratings_data_df.to_sql(name='ratings_gender', con=engine, if_exists='replace', index=True)
```
